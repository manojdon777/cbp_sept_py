```python
# Count following from string
#   1. Alphabets
#   2. Digits
#   3. Vowels
#   4. Consonents
#   5. Lower case chars
#   6. Upper case chars
#   7. Special symbols


string = 'Cdac@123'
alpha = 0
str_len = 0
digits = 0
vowels = 0
consonents = 0
lowercase = 0
uppercase = 0
special_symbol = 0
special_symbol_list = '~!@#$%^&*()+=-_,.<>;:[]{}\/? '


print(string)

for char in string:
	str_len += 1

	if char >= 'A' and char <= 'Z' or char >= 'a' and char <= 'z':
		alpha += 1

	elif char >= '0' and char <= '9':
		digits += 1


	elif char in ['a','e','i','o','u','A','E','I','O','U',]:
		vowels += 1

	elif char >= 'A' and char <= 'Z' or char >= 'a' and char <= 'z':
		if char not in ['a','e','i','o','u','A','E','I','O','U',]:
			consonents += 1

	elif char >= 'a' and char <= 'z':
		lowercase += 1

	elif char >= 'A' and char <= 'Z':
		uppercase += 1


	elif char in special_symbol_list:
		special_symbol += 1

print('String lenght =', str_len)
print('Alphabets =', alpha)
print('Digits =', digits)
print('Vowels =', vowels)
print('Consonents =', consonents)
print('Lowercase =', lowercase)
print('Uppercase =', uppercase)
print('Special symbol =', special_symbol)
```
